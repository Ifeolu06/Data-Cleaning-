```python
class_counts = dt['target'].value_counts()
class_counts
```

Visualizing the target value


```python
plt.bar(class_counts.index,class_counts.values)
plt.xlabel('target')
plt.ylabel('count')
plt.title('Bar chart of Target counts')
plt.xticks([0,1])
plt.show()
```

```python
value_counts = dt['sex'].value_counts()
value_counts
```

```python
feature = 'sex'
```

```python
plt.figure(figsize=(8,6))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of' + feature)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
age_column = 'age'
```

```python
plt.figure(figsize=(8, 6))
plt.hist(dt[age_column], bins=range(20, max(dt[age_column])+10, 10), color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=dt, x=age_column, fill=True, color='skyblue')
plt.title('Kernel Density Estimate (KDE) of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
chest_pain_column = 'chest pain type'
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=dt, x=chest_pain_column, fill=True, color='skyblue')
plt.title('Kernel Density Estimate (KDE) of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(dt[age_column], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Chest Pain Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=dt, x='age', hue='target', common_norm=False, fill=True, palette='Set2')
plt.xlabel('age')
plt.ylabel('Density')
plt.title('Age Distribution by Target')
plt.show()
```

```python
feature1= 'sex'
```

```python
feature2 = 'target'
```

```python
grouped_data = dt.groupby(feature1)[feature2].count()
```

```python
plt.figure(figsize=(10,6))
grouped_data.plot(kind='bar', color='skyblue')
plt.title('Bar Chart of ' + feature1 + 'vs Target')
plt.xlabel(feature1)
plt.ylabel('count')
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.grid(axis ='y')  # Add grid lines on the y-axis
plt.show()
```
